## Notebook Set Up

```python
from pymongo import MongoClient
import pandas as pd
from pprint import pprint
```

```python
# Create an instance of MongoClient
mongo = MongoClient(port=27017)
```

```python
# assign the uk_food database to a variable name
ukfood = mongo['uk_food']
```

```python
# review the collections in our database
print(ukfood.list_collection_names())
```

```python
# assign the collection to a variable
establishments = ukfood['establishments']
```

### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query_hygiene20 = {"scores.Hygiene": 20}

# Use count_documents to display the number of documents in the result
print(establishments.count_documents(query_hygiene20))

# Display the first document in the results using pprint
print()
pprint(establishments.find(query_hygiene20))
```

```python
# Convert the result to a Pandas DataFrame
results = establishments.find(query_hygiene20)
hygiene20 = pd.json_normalize(results)

# Display the number of rows in the DataFrame
print(hygiene20.shape)

# Display the first 10 rows of the DataFrame
hygiene20.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London in the Local Authority and has a RatingValue greater than or equal to 4.
query_london_ratingvalue_greaterthan4 = {"LocalAuthorityName": {"$regex": "London"},
        "RatingValue": {"$gte": 4}}

# Use count_documents to display the number of documents in the result
print(establishments.count_documents(query_london_ratingvalue_greaterthan4))

# Display the first document in the results using pprint
print()
pprint(establishments.find(query_london_ratingvalue_greaterthan4))
```

```python
# Convert the result to a Pandas DataFrame
results = establishments.find(query_london_ratingvalue_greaterthan4)
df = pd.json_normalize(results)

# Display the number of rows in the DataFrame
print(df.shape)

# Display the first 10 rows of the DataFrame
df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude of Penang Flavours.
# Rating value = 5
# Sort by asc hygiene score
# Top 5 establishments 

search_radius = 0.01
latitude = 51.49014200
longitude = 0.08384000

queries = {"geocode.latitude": {"$gte": latitude - search_radius, "$lte": latitude + search_radius},
        "geocode.longitude": {"$gte": longitude - search_radius, "$lte": longitude + search_radius},
        "RatingValue": 5}

sort = [("scores.Hygiene", 1)]
limit = 5

# Print the results
# Use count_documents to display the number of documents in the result
print(establishments.count_documents(queries))

# Display the first document in the results using pprint
print()
pprint(establishments.find(queries).sort(sort).limit(limit)[0])
```

```python
# Convert result to Pandas DataFrame
results = establishments.find(queries).sort(sort).limit(limit)
df = pd.json_normalize(results)

# Display the number of rows in the DataFrame
print(df.shape)

# Display the first 5 rows of the DataFrame
df.head(5)
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that:
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest
query_hygiene0 = {'$match': {'scores.Hygiene': 0
                         }
              }

group_query_local_authority = {'$group': {'_id': {
                                    'LocalAuthorityName': "$LocalAuthorityName"
                                },
                          'num_restaurants': { '$sum': 1 }, 
                         }
              }

sort_values = {'$sort': { 'num_restaurants': -1
                        }
              }

pipeline = [query_hygiene0, group_query_local_authority, sort_values]

# Print the number of documents in the result
results = list(establishments.aggregate(pipeline))

# Print the first 10 results
pprint(results[0:10])
```

```python
# Convert the result to a Pandas DataFrame
df = pd.json_normalize(results)

# Display the number of rows in the DataFrame
print(df.shape)

# Display the first 10 rows of the DataFrame
df.head(10)
```
